```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import uproot
import sys
import math
import ROOT

dir = "/Users/alexanderantonakis/Desktop/Software/AFrameAnalysis/Macros/"

sys.path.append("../Utils")
sys.path.append("../Configs")

from ChannelMap import ChannelMap

frame = 1
strip_w = 11.2 # cm 
cluster_file = "clusters_frame1.root"
config = "config_frame1.txt"
run_config = "run_config_frame1.txt"



fig_dir = "../Figs/Frame"+str(frame)+"/"

# set up the geometry of the frame
map = ChannelMap("../Configs/"+config, "../Configs/"+run_config)
map.initialize_config()
map.initialize_run_config()
map.calculate_params()
print("initialized the geometry and voltages")
print("")

# initialize the horizontal febs --> useful to have
febs = map.mac5
horiz_febs = []
for feb in febs:
    if map.is_horiz(feb):
        horiz_febs.append(feb)
        
print("All Horizontal FEBs in this file:", horiz_febs)


# Open the ROOT file and the TTree
file = uproot.open(cluster_file)  # Replace with your ROOT file
tree = file["cluster_tree"]    

# Convert the TTree to a pandas DataFrame
df = tree.arrays(library="pd")

df[:2]
```

```python
strip_df = pd.DataFrame(df['strips'].tolist())
columns = ["Run"]
for feb in horiz_febs:
    columns.append("FEB"+str(feb))
strip_df.columns = columns

print("made the strip dataframe")
strip_df[:5]
```

```python
# PLot the relative hit rate
s_test = strip_df.query("Run == 5728")
h0 = ROOT.TH1D("h0", "", len(horiz_febs), 0, len(horiz_febs))
h15 = ROOT.TH1D("h15", "", len(horiz_febs), 0, len(horiz_febs))

count = 1
for feb in horiz_febs:
    c0 = list(s_test["FEB"+str(feb)].values).count(0)
    c15 = list(s_test["FEB"+str(feb)].values).count(15)
    h0.SetBinContent(count, c0)
    h15.SetBinContent(count, c15)
    count += 1

for num in range(len(horiz_febs)):
    h0.GetXaxis().SetBinLabel(num+1, str(horiz_febs[num]))

h0.GetXaxis().SetTitle("Horizontal FEB ID")
h0.SetStats(0)
c = ROOT.TCanvas("c", "c", 700, 500)
h0.SetLineColor(2)
h15.SetLineColor(4)
h0.Draw("HISTE")
h15.Draw("HISTE Same")
c.Draw()
```

```python
time_df = pd.DataFrame(df['times'].tolist())

time_df.columns = columns

print("made the time dataframe")
time_df[:5]
```

```python

L = str(73)
R = str(159)
def dtMetric(row):
    c = row['FEB153']
    l = row['FEB'+L]
    r = row['FEB'+R]
    if l > 0 and r > 0:
        dl = abs(c - l)
        dr = abs(c - r)
        return dr - dl
    else:
        return -400
        
time_df['dt'] = time_df.apply(dtMetric, axis=1)
plt.hist(time_df.query("dt > -400")["dt"].values, bins=100, color="r", label="R = "+R+", L = "+L)
plt.xlabel(r"$\Delta$t (R - L) [ns]")
plt.ylabel("Counts")
plt.title("Frame "+str(frame)+" Cluster Candidates")
plt.yscale("log")
plt.legend()
plt.show()
```
